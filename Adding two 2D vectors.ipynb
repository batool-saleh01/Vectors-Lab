## Adding Two Vectors Together
In this part of this lab, we will plot the result of adding vector $\vec{w}$ to vector $\vec{v}$. Both vectors $\vec{v}$ and $\vec{w}$, have been defined below.

$\hspace{1cm}\vec{v} = \begin{bmatrix} 1\\ 1\end{bmatrix}$

$\hspace{1cm}\vec{w} = \begin{bmatrix} -2\\ 2\end{bmatrix}$

Below is an outline that describes what is included in the Python code to plot vector $\vec{v}$ and $\vec{w}$:

### Plotting Two Vectors
1. Make both NumPy and Matlibplot python packages available using the _import_ method 
2. Define vector $\vec{v}$ 
3. Define vector $\vec{w}$ 
4. Adjust vector settings

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vector v 
v = np.array([1,1])

# Define vector w
w = np.array([-2,2])

# Plots vector v(blue arrow) and vector w(cyan arrow) with red dot at origin (0,0) 
# using Matplotlib

# Creates axes of plot referenced 'ax'
ax = plt.axes()

# Plots red dot at origin (0,0)
ax.plot(0,0,'or')

# Plots vector v as blue arrow starting at origin 0,0
ax.arrow(0, 0, *v, color='blue', linewidth=1.5, head_width=0.15, head_length=0.15)

# Plots vector w as cyan arrow starting at origin 0,0
ax.arrow(0, 0, *w, color='cyan', linewidth=1.5, head_width=0.15, head_length=0.15)

# Sets limit for plot for x-axis
plt.xlim(-3, 4)

# Set major ticks for x-axis
major_xticks = np.arange(-3, 4)
ax.set_xticks(major_xticks)


# Sets limit for plot for y-axis
plt.ylim(-3, 4)

# Set major ticks for y-axis
major_yticks = np.arange(-3, 4)
ax.set_yticks(major_yticks)

# Creates gridlines for only major tick marks
plt.grid(visible=True, which='major')

# Displays final plot
plt.show()
```

### Vector Addition
Below we display graphically, adding vector $\vec{w}$ to vector $\vec{v}$.
1. Create vector $\vec{vw}$ by adding vector $\vec{w}$ to vector $\vec{v}$

2. Use the _ax.arrow(...)_ statement in the code below to add vector $\vec{vw}$ to the plot. Make vector $\vec{vw}$ a thicker width black colored vector.


```python
# Define vector v 
v = np.array([1,1])

# Define vector w
w = np.array([-2,2])

# Define vector vw
vw = v+w

# Creates axes of plot referenced 'ax'
ax = plt.axes()

# Plots red dot at origin (0,0)
ax.plot(0,0,'or')

# Plots vector v as blue arrow starting at origin 0,0
ax.arrow(0, 0, *v, color='blue', linewidth=2, head_width=0.15, head_length=0.15)

# Plots vector w as cyan arrow with origin defined by vector v
ax.arrow(v[0], v[1], *w, linestyle='dotted', color='cyan', linewidth=2, head_width=0.15, head_length=0.15)

# Plots vector vw as black arrow with origin
ax.arrow(0, 0, *vw, color='black', linewidth=2, head_width=0.15, head_length=0.15)

# Sets limit for plot for x-axis
plt.xlim(-3, 4)
major_xticks = np.arange(-3, 4)
ax.set_xticks(major_xticks)


# Sets limit for plot for y-axis
plt.ylim(-3, 4)
major_yticks = np.arange(-3, 4)
ax.set_yticks(major_yticks)

# Creates gridlines for only major tick marks
plt.grid(visible=True, which='major')

# Displays final plot
plt.show()
```
